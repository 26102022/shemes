```python
wels=5
rows=3
list1=[0]*wels*rows
for i in range(wels-1):
    list1[i]=(round(random.gauss(42, 2.1),1))
for i in range(rows*wels-wels):
    list1[wels+i]=(round(random.gauss(25, 2.1),1))
print(list1)
sum=[0]
mat=[]
maat=[]
for i in range(wels-1):
    sum.append(sum[i]+list1[i])

for i in range(wels):
    sum.append(sum[i]+list1[i+wels])

for i in range(wels):
    sum.append(sum[wels+i]+list1[i+wels*2])
for i in range(3):
    mat=[]
    for j in range(5):
        mat.append(sum[i*wels+j])
    maat.append(mat)
for j in range(rows):
    for i in range(wels):
        maat[j][i]=float(maat[j][i])+float(round(random.gauss(500, 10.5),1))

mt=[[],[],[]]
mt[0]=maat[2]
mt[1]=maat[1]
mt[2]=maat[0]
fig, ax1 = plt.subplots()
ax1.matshow(mt)
for (i, j), z in np.ndenumerate(mt):
    ax1.text(j, i, '{:0.1f}'.format(z), ha='center', va='center',
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='0.3'))
aa=[]
ss=[]
ss.append(mt[rows-1][0]-mt[rows-1][0])
ss.append(mt[rows-2][0]-mt[rows-1][0])
ss.append(mt[0][0]-mt[rows-2][0])
aa.append(ss)

for i in range(wels-1):
    ss=[]
    ss.append(mt[rows-1][i+1]-mt[rows-1][i])
    ss.append(mt[rows-2][i+1]-mt[rows-1][i+1])
    ss.append(mt[0][i+1]-mt[rows-2][i+1])
    aa.append(ss)


maat=np.transpose(aa)
mt[0]=maat[2]
mt[1]=maat[1]
mt[2]=maat[0]
fig, ax = plt.subplots()
aaaaa=ax.matshow(mt)

for (i, j), z in np.ndenumerate(mt):
    ax.text(j, i, '{:0.1f}'.format(z), ha='center', va='center',
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='0.3'))
plt.show()
```
