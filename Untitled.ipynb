```python
from collections import Counter
import seaborn as sns
list2=[]
list1=[]
for i in range(10000):
    list2.append(round(random.gauss(25,0.8),1))
for i in range(10000):
    list1.append(round(random.gauss(25,0.8),1))
print((list1))
print((list2))
a = list1
b = list2

c = []

for item in a:
    if item not in c:
        a_item = a.count(item)
        b_item = b.count(item)
        min_count = min(a_item, b_item)
        # c += [item] * min_count
        for i in range(min_count):
            c.append(item)

print(len(c)/10000)
```
